## NEURAL NETWORK FLOW

```text
1. Input Image
   (Image converted into pixel values)

2. Flatten
   (2D image converted into 1D vector)

3. Linear Layer
   (Learns patterns using weights)

4. ReLU Activation
   (Removes negative values)

5. Hidden Features
   (Detects edges, curves, shapes)

6. Final Output Layer
   (Generates class scores/logits)

7. Softmax
   (Converts scores into probabilities)

8. Prediction
   (Highest probability becomes final answer)
```
